## And now for some fun - an adversarial conversation between Chatbots..

You're already familar with prompts being organized into lists like:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "user prompt here"}
]
```

In fact this structure can be used to reflect a longer conversation history:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "first user prompt here"},
    {"role": "assistant", "content": "the assistant's response"},
    {"role": "user", "content": "the new user prompt"},
]
```

And we can use this approach to engage in a longer interaction with history.

````python
# Let's make a conversation between GPT-4.1-mini and Claude-haiku-4.5
# We're using cheap versions of models so the costs will be minimal

import google.generativeai as genai
from groq import Groq
from IPython.display  import Markdown, display 

genai.configure(api_key = google_api_key)
groq_client = Groq(api_key = groq_api_key)

gemini_model = "gemini-2.5-flash"
groq_model = "llama-3.3-70b-versatile" 

groq_system = """You are a chatbot who is very argumentative;
you disagree with anything in the conversation and you challenge everything, in a snarky way."""

gemini_system = """You are a very polite, courteous chatbot.
You try to agree with everything the other person says, or find common ground.
If the other person is argumentative, you try to calm them down and keep chatting."""

groq_messages = ["Hi there"]
gemini_messages = ["Hi"]
````

````python
gemini_model_instance = genai.GenerativeModel(
    model_name=gemini_model,
    system_instruction=gemini_system
)

gemini_chat = gemini_model_instance.start_chat(history=[])
````

````python
def call_groq():
    messages = [{"role": "system", "content": groq_system}]
    
    for g, gm in zip(groq_messages, gemini_messages):
        messages.append({"role": "assistant", "content": g})
        messages.append({"role": "user", "content": gm})

    try:
        response = groq_client.chat.completions.create(
            model=groq_model,
            messages=messages,
            temperature=0.7
        )
        return response.choices[0].message.content

    except Exception as e:
        return f"[Groq Error]: {str(e)}"
````

````python
call_groq()
````

````python
import time 
def call_gemini():
    try:
        response = gemini_chat.send_message(groq_messages[-1])
        return response.text

    except Exception as e:
        # simple retry mechanism
        time.sleep(3)
        try:
            response = gemini_chat.send_message(groq_messages[-1])
            return response.text
        except Exception as e:
            return f"[Gemini Error]: {str(e)}"
````

````python
call_gemini()
````

````python
call_groq()
````

````python
groq_messages = ["Hi there"]
gemini_messages = ["Hi"]

display(Markdown(f"### Groq:\n{groq_messages[0]}\n"))
display(Markdown(f"### Gemini:\n{gemini_messages[0]}\n"))

for i in range(5):
    # 🔴 GROQ TURN
    groq_next = call_groq()
    display(Markdown(f"### Groq:\n{groq_next}\n"))
    groq_messages.append(groq_next)

    time.sleep(1.5)  # prevent rate limit

    # 🔵 GEMINI TURN
    gemini_next = call_gemini()
    display(Markdown(f"### Gemini:\n{gemini_next}\n"))
    gemini_messages.append(gemini_next)

    time.sleep(1.5)
````

## Additional exercise

You could also try replacing one of the models with an open source model running with Ollama.

````python
import time
import google.generativeai as genai 
from groq import Groq 
from IPython.display import Markdown, display 

genai.configure(api_key = google_api_key)
groq_client = Groq(api_key = groq_api_key) 

GROQ_MODEL = "llama-3.3-70b-versatile"
GEMINI_MODEL = "gemini-1.5-flash-8b" 

alex_system = """You are Alex, a chatbot who is very argumentative; 
you disagree with anything and challenge everythong in a snarky way. You are in conversation with Blake and Charlie."""

blake_system = """You are Blake, a polite and calm chatbot. You try to agree and de-escalate arguments. You are in a conversation with Alex and Charlie"""  

charlie_system = """You are Charlie, a logical and neutral thinker. You balance both sides and give rational opinions. You are in a conversation with Alex and Blake."""  

conversation = """Alex: Hi  there
Blake: Hi
Charlie: Hello everyone"""  


display(Markdown(f"### Conversation Start: \n{conversation}")) 
````

````python
def call_groq(system_prompt, name, conversation):
    user_prompt = f"""You are {name}, in conversation with others. 
    The conversation so far: {conversation}
    Now respond as {name}. 
    """  

    response =  groq_client.chat.completions.create(
        model = GROQ_MODEL,
        messages =  [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt} 
        ],
        temperature = 0.7
    )

    return response.choices[0].message.content
````

````python
import time

def call_gemini(system_prompt, name, conversation, max_retries=3):
    model = genai.GenerativeModel(
        model_name="GEMINI_MODEL",  # lighter = fewer errors
        system_instruction=system_prompt
    )

    user_prompt = f"""
You are {name}, in conversation with others.

The conversation so far:
{conversation}

Now respond as {name}.
"""

    for attempt in range(max_retries):
        try:
            response = model.generate_content(user_prompt)
            return response.text

        except Exception as e:
            wait = 2 ** attempt  # exponential backoff
            print(f"[Gemini retry {attempt+1}] waiting {wait}s...")
            time.sleep(wait)

    return "[Gemini failed after retries]"
````

````python
for i in range(3):
    alex_reply = call_groq(alex_system, "Alex", conversation)
    conversation += f"Alex: {alex_reply}\n"
    display(Markdown(f"### Alex:\n{alex_reply}\n"))
    time.sleep(2.5) 

    blake_reply = call_gemini(blake_system, "Blake", conversation)

    if "failed" in blake_reply.lower():
        blake_reply = call_groq(blake_system, "Blake", conversation)
    
    conversation += f"Blake: {blake_reply}\n"
    display(Markdown(f"### Blake:\n{blake_reply}\n"))

    charlie_reply = call_groq(charlie_system, "Charlie", conversation)
    conversation += f"Charlie: {charlie_reply}\n"
    display(Markdown(f"### Charlie:\n{charlie_reply}\n"))
    time.sleep(1.5)

    conversation = conversation[-3000:]
````
