## Hugging Face Libraries

1. hub - log in through code and connect to the hugging face platform.

2. datasets - first use the hub library to connect to the hub and then download a dataset. It then becomes a part of this datasets library. Each object represents a dataset.

3. transformers - to run a model, train it and do any sort of experimenting with it.

4. peft (Parameter Efficient Fine Tuning) - to efficiently train a model by yourself without tweaking any number of parameters that will there in a model. It uses LoRA (Low-Rank Adaption) which efficiently adapts large pre-trained models by freezing original weights and training small "adapter" matrices, often reducing trainable parameters by over 99%.

5. trl (Transformers Reinforcement Learning) - to train transformers.

6. accelerate - allows you to distribute models across multiple GPUs.
