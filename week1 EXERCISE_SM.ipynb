```python
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display 
from openai import OpenAI 
```

```python
# API KEYS LOAD
load_dotenv(override = True)

groq_api_key = os.getenv("GROQ_API_KEY")
google_api_key = os.getenv("GOOGLE_API_KEY")

if not groq_api_key:
    raise ValueError("Missing GROQ_API_KEY")

if not google_api_key:
    raise ValueError("Missing GOOGLE_API_KEY")
```

```python
# constants
MODEL_GEMINI = "gemini-3-flash"
MODEL_GROQ = "llama-3.3-70b-versatile"
```

```python
# set up environment for groq
groq = OpenAI(
    base_url = "https://api.groq.com/openai/v1",
    api_key = groq_api_key
)
```

```python
# set up environment for gemini
gemini = OpenAI(
    base_url = "https://generativelanguage.googleapis.com/v1beta/openai/",
    api_key = google_api_key
)
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get groq to answer, with streaming
def ask_groq(question):
    response = groq.chat.completions.create(
        model = MODEL_GROQ,
        messages = [
            {"role": "system", "content": "You are a technical tutor."},
            {"role": "user", "content": question}
        ],
        temperature = 0.3
    )

    return response.choices[0].message.content
```

```python
print("GROQ Answer:\n")
groq_answer = ask_groq(question)
display(Markdown(groq_answer))
```

```python
# Get gemini to answer
# Get groq to answer, with streaming
def ask_gemini(question):
    response = gemini.chat.completions.create(
        model = MODEL_GEMINI,
        messages = [
            {"role": "system", "content": "You are a technical tutor."},
            {"role": "user", "content": question}
        ],
        temperature = 0.3
    )

    return response.choices[0].message.content
```

```python
# usinng ollama
import ollama

def ask_ollama(question):
    response = ollama.chat(
        model = "deepseek-r1:1.5b",
        messages = [
            {"role": "system", "content": "you are a technical tutor"},
            {"role": "user", "content": question}
        ]
    )
    return response['message']['content']
```

```python
print("OLLAMA ANSWER:\n")
ollama_answer = ask_ollama(question)
display(Markdown(ollama_answer))
```
